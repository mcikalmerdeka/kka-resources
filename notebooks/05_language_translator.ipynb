## üìö STEP 3: Core Function
This function sends your message to the AI and gets a response back

```python
# Function to translate text
def translate_text(text, source_lang, target_lang, include_cultural_notes):
    """Translate text and optionally include cultural notes"""
    cultural_note = "Sertakan juga catatan budaya atau konteks menarik tentang terjemahan ini." if include_cultural_notes else ""
    
    prompt = f"""Terjemahkan teks berikut dari {source_lang} ke {target_lang}:

"{text}"

{cultural_note}
"""
    
    try:
        response = requests.post(
            f"{_BACKEND}/chat/middle",
            headers={"X-API-Key": _KEY},
            json={"prompt": prompt},
            timeout=30
        )
        response.raise_for_status()
        return response.json()["response"]
    except Exception as e:
        return f"Error: {str(e)}"
```

## üìö STEP 4: Create Interface

Set up the chat interface with Gradio

```python
languages = ["Bahasa Indonesia", "English", "Mandarin", "Arabic", "Spanish", "French", "Japanese", "Korean"]

interface = gr.Interface(
    fn=translate_text,
    inputs=[
        gr.Textbox(label="Text to Translate", lines=4, placeholder="Enter text here..."),
        gr.Dropdown(languages, label="From", value="English"),
        gr.Dropdown(languages, label="To", value="Bahasa Indonesia"),
        gr.Checkbox(label="Include cultural notes", value=False)
    ],
    outputs=gr.Textbox(label="Translation", lines=6),
    title="üåç Language Translator",
    description="Translate between languages and learn cultural insights!",
    examples=[
        ["Hello, how are you?", "English", "Bahasa Indonesia", False],
        ["Selamat pagi, apa kabar?", "Bahasa Indonesia", "English", True]
    ]
)
```
