## ðŸ“š STEP 3: Core Function
This function sends your message to the AI and gets a response back

```python
# Function to generate story
def generate_story(character, setting, theme):
    """Generate a creative story based on user inputs"""
    prompt = f"""Buatkan cerita pendek yang menarik dengan:
    - Karakter utama: {character}
    - Setting: {setting}
    - Tema: {theme}
    
    Cerita harus menarik, kreatif, dan sesuai untuk anak-anak."""
    
    try:
        response = requests.post(
            f"{_BACKEND}/chat/elementary",
            headers={"X-API-Key": _KEY},
            json={"prompt": prompt},
            timeout=30
        )
        response.raise_for_status()
        return response.json()["response"]
    except Exception as e:
        return f"Error: {str(e)}"
```

## ðŸ“š STEP 4: Create Interface

Set up the chat interface with Gradio

```python
interface = gr.Interface(
    fn=generate_story,
    inputs=[
        gr.Textbox(label="Karakter Utama", placeholder="e.g., Seekor kelinci pemberani"),
        gr.Textbox(label="Setting/Tempat", placeholder="e.g., Hutan ajaib"),
        gr.Dropdown(
            ["Petualangan", "Persahabatan", "Keberanian", "Misteri", "Sains"],
            label="Tema Cerita"
        )
    ],
    outputs=gr.Textbox(label="Cerita Anda", lines=15),
    title="ðŸ“– Story Generator",
    description="Buat cerita kreatif dengan AI! Pilih karakter, tempat, dan tema.",
    examples=[
        ["Robot kecil", "Kota masa depan", "Persahabatan"],
        ["Anak petani", "Desa pegunungan", "Keberanian"],
        ["Kucing detektif", "Perpustakaan tua", "Misteri"]
    ]
)
```
