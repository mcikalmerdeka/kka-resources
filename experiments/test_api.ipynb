```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

# Deployed backend URL
API_URL = os.getenv("BACKEND_API_URL")

# Define the function to ask the AI
def ask_ai(question, level="elementary"):
    """
    level: "elementary", "middle", or "highschool"
    """
    response = requests.post(
        f"{API_URL}/chat/{level}",
        params={"prompt": question}
    )
    return response.json()["response"]
```

```python
# First test
question = "Apa itu kecerdasan artifisial dan bagaimana cara kerjanya secara sederhana?"
answer = ask_ai(question, level="elementary")
print(answer)
```

```python
# Second test
question = "Bagaimana cara kerja sebuah model regresi linear?"
answer = ask_ai(question)
print(answer)
```

```python
# Third test
question = "Apa aplikasi dari kecerdasan artifisial untuk pembuatan game?"
answer = ask_ai(question, level="highschool")
print(answer)
```
