## Analyze each reference files

```python
from pathlib import Path

# List all files in the reference folder
reference_folder = Path(r"E:\Koding dan Kecerdasan Artificial\reference")
files = list(reference_folder.iterdir())

print("Files in reference folder:")
for file in files:
    print(f"  {file.name}")
```

```python
import pypdf
import warnings

# Suppress pypdf warnings about malformed PDFs
warnings.filterwarnings("ignore", category=UserWarning, module="pypdf")

# temporary: exclude these 2 files
files_to_exclude = [
    "Buku Putih Peta Jalan KA Nasional.pdf",
    "Konsep Pedoman Etika.pdf"]

# Filter out the files to exclude
file_paths = [file for file in files if file.suffix == '.pdf' and file.name not in files_to_exclude]

for file_path in file_paths:
    # Print the file name
    print(f"Processing file: {file_path.name}")

    # Open the PDF file
    with open(file_path, 'rb') as file:
        pdf_reader = pypdf.PdfReader(file)
        text_content = ""
        
        # Extract text from each page
        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text_content += page.extract_text()

    # Print the document information
    print(f"Document Information for {file_path.name}:")    
    print(f"Total pages: {len(pdf_reader.pages)}")
    print(f"Total characters: {len(text_content)}")

    # Print a separator
    print("===="*50)
    print()
```
